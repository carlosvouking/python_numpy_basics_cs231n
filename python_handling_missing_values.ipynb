```python
import pandas as pd
import numpy as np
```

```python
# Null values in data_set.csv as for pandas.
df_isnull_check = df.isnull()
df_isnull_check
```

```python
def cell_highlight(val):
    color = 'orangered' if val == 'True'  else 'green'
    return 'color: {color}.'
```

```python
df_isnull_check.style.applymap(cell_highlight)
```
